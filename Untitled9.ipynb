```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from osgeo import gdal
import os
df = pd.read_excel( r'F:\SUKI.xls')                     
x_data=df.loc[:,['HV','VH','HH','VV']]
y_data=df.loc[:,'AGB']
def R2(y_true, y_pred):  
    # 定义R2 ,真值在前
    return 1 - ((y_pred - y_true)**2).sum() / ((y_true - y_true.mean())**2).sum()
e=50
E=[]
trR2=[]
teR2=[]
x_train, x_test, y_train, y_test = train_test_split(
        x_data, y_data, test_size=0.15,random_state =121)
RFR = RandomForestRegressor(bootstrap=True, criterion='mse', 

       max_features='auto', max_leaf_nodes=40,max_depth=10,

       min_samples_leaf=1,random_state =121,

       min_samples_split=2, min_weight_fraction_leaf=0.0,

       n_estimators=400,n_jobs = 2, oob_score=True)

model = RFR.fit(x_train, y_train)
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
b1=0.9
b2=0.2
b3=0.2
b4=0.4
b5=0.2

aa=np.array([b1,b2,b3,b4]).reshape(1, -1)
bb=model.predict(aa)
mse = mean_squared_error(y_train, y_train_pred)
print("MSE: %.6f" % mse)
t1=R2(y_train,y_train_pred)
t2=R2(y_test,y_test_pred)
print(t1)
print(t2)
print(bb[0])
print('Over')




#随机森林AGB栅格应用
from osgeo import gdal
dataset1 = gdal.Open(r'F:\TEST\VH_BS1.tif')  # 打开文件
im_width = dataset1.RasterXSize  # 栅格矩阵的列数
im_height = dataset1.RasterYSize  # 栅格矩阵的行数
im_geotrans = dataset1.GetGeoTransform()  # 仿射矩阵
im_proj = dataset1.GetProjection()  # 地图投影信息
im_band = dataset1.GetRasterBand(1)

def array2raster(outpath, array, geoTransform, proj):
    cols = array.shape[1]
    rows = array.shape[0]
    driver = gdal.GetDriverByName('Gtiff')
    outRaster = driver.Create(outpath, cols, rows, 1, im_band.DataType)
    outRaster.SetGeoTransform(geoTransform)  # 参数2,6为水平垂直分辨率，参数3,5表示图片是指北的
    outband = outRaster.GetRasterBand(1)
    outband.WriteArray(array)
    outRaster.SetProjection(proj)  # 将几何对象的数据导出为wkt格式
    outRaster.FlushCache()

im_DATA = dataset1.ReadAsArray(0, 0, im_width, im_height)  # 将数据写成数组，对应栅格矩阵
print(dataset1.GetRasterBand(1).GetNoDataValue())  #获取栅格影像NoData值
ND = dataset1.GetRasterBand(1).GetNoDataValue() #  获取栅格NoData值

print('%f'%ND)  #打印栅格NoData全部位数的值
da_HV =gdal.Open((r'F:\TEST\HV_BS1')+ '.tif')  # 打开文件
da_HV
```
